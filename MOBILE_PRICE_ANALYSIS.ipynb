**MOBILE PRICE ANALYSIS**



```python
#Loading Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Reading the datafile
df = pd.read_csv("Mobiles Dataset (2025).csv", encoding="latin-1")
#To check the data(rows,cols)
df.shape
```

```python
#Converting All Country Currency to Indian Rupee
# Exchange rates relative to INR (1 unit in local currency = X INR)
exchange_rates_inr = {
    'PKR': 0.30,   # 1 PKR ≈ 0.30 INR
    'INR': 1.0,    # 1 INR = 1 INR
    'CNY': 11.6,   # 1 CNY ≈ 11.6 INR
    'USD': 83.0,   # 1 USD ≈ 83 INR
    'AED': 22.6    # 1 AED ≈ 22.6 INR
}

# Function to convert to INR
def convert_to_inr(value):
    try:
        if isinstance(value, str):
            parts = value.strip().replace(',', '').split()
            if len(parts) == 2:  # Expected format like "USD 500"
                curr, amount = parts
                rate = exchange_rates_inr.get(curr)
                if rate:
                    return float(amount) * rate
        return None
    except:
        return None

# Columns to process
columns_to_change = [
    'Launched Price (Pakistan)', 'Launched Price (India)',
    'Launched Price (China)', 'Launched Price (USA)',
    'Launched Price (Dubai)'
]

# Apply conversion
for col in columns_to_change:
    df[col + ' (INR)'] = df[col].apply(convert_to_inr)

# Drop old columns if not needed
df = df.drop(columns=columns_to_change)
```

```python
#Removing NULL values
df.dropna(inplace=True)
```

```python
# Calculate average price per year in India
avg_price_per_year = df.groupby("Launched Year")["Launched Price (India) (INR)"].mean()

print(avg_price_per_year)
```

```python
plt.figure(figsize=(10,5))

# Scatter plot of average prices
sns.scatterplot(x=avg_price_per_year.index,
               y=avg_price_per_year,
               label='Avg price',
               color='Green')

# Fit a polynomial trend line (degree=2)
z = np.polyfit(avg_price_per_year.index, avg_price_per_year, 2)
p = np.poly1d(z)

# Plot trend line
plt.plot(avg_price_per_year.index, p(avg_price_per_year.index),
         color='purple', linestyle="--", label="Trend Line")

plt.title("Trend of Mobile Phone Prices in India Over Years")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select only price columns
price_columns = [
    'Launched Price (Pakistan) (INR)',
    'Launched Price (India) (INR)',
    'Launched Price (China) (INR)',
    'Launched Price (USA) (INR)',
    'Launched Price (Dubai) (INR)'
]

# Find most expensive phone in each country
most_expensive_models = {}
for col in price_columns:
    max_row = df.loc[df[col].idxmax()]
    most_expensive_models[col] = {
        'Model Name': max_row['Model Name'],
        'Company Name': max_row['Company Name'],
        'Price (INR)': max_row[col]
    }

# Convert to DataFrame for easy plotting
import pandas as pd
most_expensive_df = pd.DataFrame(most_expensive_models).T

# Plot
plt.figure(figsize=(10,6))
plt.bar(most_expensive_df.index, most_expensive_df['Price (INR)'], color='skyblue')

# Add labels (model + company) on bars
for i, val in enumerate(most_expensive_df['Price (INR)']):
    plt.text(i, val + 2000,
             f"{most_expensive_df.iloc[i]['Company Name']} {most_expensive_df.iloc[i]['Model Name']}",
             ha='center', rotation=90, fontsize=8)

plt.title("Most Expensive Mobile Model in Each Country (INR)")
plt.ylabel("Price (INR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Average Price by country
avg_price_by_country = df[
    ["Launched Price (Pakistan) (INR)",
     "Launched Price (India) (INR)",
     "Launched Price (China) (INR)",
     "Launched Price (USA) (INR)",
     "Launched Price (Dubai) (INR)"]
].mean().sort_values(ascending=False)
avg_price_by_country
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_price_by_country.index,
            y=avg_price_by_country.values,
            palette="viridis")

plt.xlabel("Country")
plt.ylabel("Average Launched Price (INR)")
plt.title("Comparison of Mobile Prices")
plt.xticks(rotation=20)
plt.show()
```

```python
#COMPANY WISE MOBILE PRICES ACROSS COUNTRIES(5)
import pandas as pd

# Price columns
price_columns = [
    'Launched Price (Pakistan) (INR)',
    'Launched Price (India) (INR)',
    'Launched Price (China) (INR)',
    'Launched Price (USA) (INR)',
    'Launched Price (Dubai) (INR)'
]

# Group by company and calculate average price in each country
company_avg_prices = df.groupby('Company Name')[price_columns].mean().reset_index()
company_avg_prices
```

```python
from matplotlib.colors import LinearSegmentedColormap

# Create custom colormap
custom_cmap = LinearSegmentedColormap.from_list("green_violet", ["lightgreen", "violet"])

plt.figure(figsize=(10,8))
sns.heatmap(company_avg_prices, annot=True, fmt=".0f", cmap=custom_cmap)
plt.title("Company-wise Mobile Prices Across 5 Countries (INR)")
plt.ylabel("Company")
plt.xlabel("Country")
plt.show()
```

```python
#Nokia has a significant outlier in the USA , much higher than other regions.This may be a data entry issue or an extreme pricing case.
#Apple, Sony, and Huawei have the highest prices across all regions.
#Pakistan generally has the lowest prices, while India and the USA have higher prices.
#Budget brands like Infinix, Tecno, and Lenovo have consistently low prices across all countries.
```

```python
#Avg weight/company
# Remove 'g' and any extra characters, then convert to float
df['Mobile Weight'] = df['Mobile Weight'].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
```

```python
# Group by company and calculate average weight
avg_weight_company = df.groupby('Company Name')['Mobile Weight'].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(x=avg_weight_company.index, y=avg_weight_company.values, palette="PuBuGn")
plt.xticks(rotation=45)
plt.title("Average Mobile Weight by Company")
plt.xlabel("Company")
plt.ylabel("Average Weight (g)")
plt.show()
```

```python
#Avg battery /Mobile
# Clean Battery Capacity column
df['Battery Capacity'] = df['Battery Capacity'].astype(str).str.replace(',', '').str.replace('mAh', '')
df['Battery Capacity'] = df['Battery Capacity'].astype(float)
avg_battery_company = df.groupby('Company Name')['Battery Capacity'].mean().sort_values(ascending=False)
```

```python
#visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(x=avg_battery_company.index, y=avg_battery_company.values, palette="cividis")
plt.xticks(rotation=45)
plt.title("Average Battery Capacity by Mobile Company")
plt.xlabel("Company")
plt.ylabel("Average Battery Capacity (mAh)")
plt.show()
#milliampere-hour.
```
